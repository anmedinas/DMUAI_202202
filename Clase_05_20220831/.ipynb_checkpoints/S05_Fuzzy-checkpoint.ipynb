```python
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt 
from plotnine import *
from sklearn import datasets
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import seaborn as sns
import warnings 

# Omite Warnings 
warnings.simplefilter("ignore")

# Configuracion Entorno Plot
sns.set(rc={'figure.figsize':(5.27,5.27)})
```

```python
wine = datasets.load_wine()
scaler = StandardScaler() 
scaled_features = scaler.fit_transform(wine.data)
```

### **Fuzzy C-Means**

`Sklean` no tiene implementada esta función aún. Sin embargo, utilizaremos para ese proposito `skfuzzy` el cuál retorna múltiples valores. El único problema de esta implementación, es que considera las filas como variables y las columnas como puntos, por lo que antes de ingresar los datos, hay que trasponerlos (`data.transpose()`).

La función, se debe instalar y posteriormente cargarla. Por otro lado, la sintaxis para invocar a `cmeans` es:

```python
from IPython.display import Image
Image('fuzzy.png')
```

Parámetros:

* `data`: Data a conglomerar.
* `c`: Número de clusters.
* `m`: parámetro difusor.
* `error`: Criterio de parada según la tolerancia del error.
* `maxIter`: Número máximo de iteraciones.

Valores de retorno:
* `cntr`: Centro de los clusters.
* `u`: Matriz de pertenencia.
* `u0`: Matriz de pertenencia inicial.
* `d`: Matriz final de la distancia euclidiana.
* `jm`: Log de la función objetivo.
* `p`: Número de iteraciones ejecutadas.
* `fpc`: Coeficiente de partición final. Varía entre  0 y 1. Valores cercanos a 1, son mejores.

```python
import skfuzzy as fuzzy
(CentrosClusters, u, _, _, _, _, fpc) = fuzzy.cluster.cmeans(scaled_features.transpose(), c = 4, m = 2.3, error= 0.005, maxiter= 1000)
```

```python
print(CentrosClusters)
```

```python
print("\nProbabilidad de pertenencia para el primer punto")
print(u[:,0])
```

```python
clusters = np.argmax(u, axis=0)
print(clusters)
```

```python
print(fpc)
```

```python
sse = []
for k in range(1, 11):
    (CentrosClusters, u, _, _, _, _, fpc) = fuzzy.cluster.cmeans(scaled_features.transpose(), c=k, m=2, error=0.0005, maxiter=1000)
    sse.append(fpc)
```

```python
sse
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1,11), sse, marker='o',linestyle='--')
plt.title('Fuzzy C-Mean',fontsize=15)
plt.xlabel('Número de Clusters',fontsize=13)
plt.ylabel('fpc', fontsize=13)
plt.show()
```

```python
sse = []
for k in range(1, 11):
    (CentrosClusters, u, _, _, _, _, fpc) = fuzzy.cluster.cmeans(scaled_features.transpose(), c = k, m = 2, error = 0.0005, maxiter = 1000)
    clusters = np.argmax(u, axis=0)
    tempWCD = 0
    for j in range(CentrosClusters.shape[0]):
        tempWCD += np.square(scaled_features[clusters==j,]-CentrosClusters[j,:]).sum()
    sse.append(tempWCD)
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1,11), sse, marker='o',linestyle='--')
plt.axvline(x=3, color='r', linestyle='--')
plt.title('Fuzzy C-Mean',fontsize=15)
plt.xlabel('Número de Clusters',fontsize=13)
plt.ylabel('fpc', fontsize=13)
plt.show()
```

```python
(CentrosClusters, u, _, _, _, _, fpc) = fuzzy.cluster.cmeans(scaled_features.transpose(), c = 3, m = 1.5, error = 0.0005, maxiter=1000)
clusters = np.argmax(u, axis=0)

pca = PCA(n_components = 2)
pca.fit(scaled_features)

tmpData = pca.transform(scaled_features)
tmpData = pd.DataFrame(tmpData,columns=["PC1","PC2"])
tmpData["labels"]=clusters
tmpData.insert(3, "colorPoint","")
for i in range(tmpData.shape[0]):
    output=(((u[:,i])*255).astype("int"))
    tmpData.iloc[i,3]='#%02x%02x%02x' % (output[0],output[1],output[2])
```

```python
tmpData.head(4)
```

```python
(ggplot(tmpData) + aes(x="PC1",y="PC2") + theme_bw()+
   geom_point(show_legend = False, color = tmpData.colorPoint)
)
```

```python
colors = ['b', 'orange', 'g', 'r', 'c', 'm', 'y', 'k', 'Brown', 'ForestGreen']
xpts = tmpData['PC1']
ypts = tmpData['PC2']

fig1, axes1 = plt.subplots(3, 3, figsize=(10, 8))
alldata = np.vstack((xpts, ypts))
fpcs = []

for ncenters, ax in enumerate(axes1.reshape(-1), 2):
    cntr, u, u0, d, jm, p, fpc = fuzzy.cluster.cmeans(
        alldata, ncenters, 2, error=0.005, maxiter=1000, init=None)

    # Store fpc values for later
    fpcs.append(fpc)

    # Plot assigned clusters, for each data point in training set
    cluster_membership = np.argmax(u, axis=0)
    for j in range(ncenters):
        ax.plot(xpts[cluster_membership == j],
                ypts[cluster_membership == j], '*', color=colors[j],markersize=8)

    # Mark the center of each fuzzy cluster
    for pt in cntr:
        ax.plot(pt[0], pt[1], 'rs', markersize=10)

    ax.set_title('Centers = {0}; FPC = {1:.2f}'.format(ncenters, fpc))
    ax.axis('off')

fig1.tight_layout()
```
